```python
    
for url_id in url_arr:
    n_pages = 0
    # This for gets us into next page after every iteration.
    for page in range(1,num_of_pages_to_scrape):  #The no. of pages you wanted to scrape considering pages available in the url.
        n_pages += 1
        if((page == 1) | (page==2) ):
            url=url_id
        else:
            url=url_id+'?page='+'str(page)'        
        
        res = requests.get(url) # Here res is response data.

        page_html = soup(res.text, 'html.parser') 
        house_containers = page_html.find_all("li",class_="cardholder") # Class containing overall data of a property.
        for data in house_containers:
            type=data.find_all("a",class_="typelink") # Class containing title name of a property.
            for i in type:
                d=i.text
                title.append(d)
                city.append(default_city_name)
            
            location=data.find_all("a",class_="loclink")
            for i in location:
                loc = i.find("span",itemprop="addressLocality").text
                locationnew.append(loc)
            
            cost=data.find_all("td",class_="price")            
            for i in cost:
                d=i.text
                if 'L' in d:
                    a = d.split()[0]
                    price.append(float(a)) # Type conversion of obtained data.
                elif 'Cr' in d:
                    b = d.split()[0]
                    price.append(float(b) * 100) # Coverting CRORES into LAKHS.
                else:  # If no value found in the class return value 0.
                    e=0
                    price.append(float(e)) 

            rate_sqft=data.find_all('td',class_="lbl rate")            
            for i in rate_sqft:
                d=i.text
                res= d.split('/',)[0]
                r=re.sub(",","" ,res) # Substitute no character in the place of ,(comma).
                price_per_sqft.append(r)

            area_sqft=data.find_all('td',class_="size")            
            for i in area_sqft:
                d=i.text
                area_in_sqft.append(d)

            building_status_=data.find_all("td",class_="val")            
            for i in building_status_:
                d=i.text
                building_status.append(d)

            try:
                attr_dict_1 = data.find("div",class_="cardLayout clearfix").attrs
                sub_page_link = attr_dict_1['itemid']
                #print("sub_page_link : ", sub_page_link)
                try:
                    sub_page_resp = requests.get(sub_page_link)
                    #time.sleep(5)               # Intentional Delay
                    sub_page_html = soup(sub_page_resp.text, 'html.parser')
                    last_updated_date_ = sub_page_html.find('div', class_='updated-date')
                    #print("last_updated_date_" , last_updated_date_)
                    for i in last_updated_date_:
                        d=i.text
                        last_updated_date.append(d)
                except Exception as inner_e:
                    #print("in inner exception block ", inner_e.__class__)
                    pass
            except Exception as outer_e:
                #print("in outer exception block ", outer_e.__class__)
                pass
```
